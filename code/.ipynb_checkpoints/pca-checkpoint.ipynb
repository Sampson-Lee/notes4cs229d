### 第一部分
使用sklearn的api，参考[scikit-learn的PCA](http://blog.csdn.net/u012162613/article/details/42192293)

```python
import numpy as np

data = np.array([[0,10,2],[0,10,4],[0,10,6]])
print(data)  # 降维前的样本
```

```python
from sklearn.decomposition import PCA

data_pca = PCA(n_components=1,copy=True,random_state=8)
newdata = data_pca.fit_transform(data)
print(newdata)  #降维后的样本
```

```python
print(pca.components_) # 投影方向向量
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()

X = iris.data
y = iris.target
print(X.shape)
print(y.shape)
```

```python
iris_pca = PCA(n_components=2,copy=False,random_state=8)

X = iris_pca.fit_transform(X)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)  # scatter绘制散点图，绘制X两个维度数据，颜色与y一致
plt.show()
print(iris_pca.components_) # 两个投影方向上的向量
```

### 第二部分
直接写代码，深入理解算法过程，参考[PCA算法实现python](http://blog.csdn.net/u012162613/article/details/42177327)
步骤如下：

1）将原始数据按列组成n行m列矩阵X

2）将X的每一行（代表一个属性字段）进行零均值化，即减去这一行的均值

3）求出协方差矩阵C=1/mXX^T

4）求出协方差矩阵的特征值及对应的特征向量

5）将特征向量按对应特征值大小从上到下按行排列成矩阵，取前k行组成矩阵P

6）Y=PX即为降维到k维后的数据

有关SVD的分析详见[奇异值分解 SVD 的数学解释](http://blog.csdn.net/u010099080/article/details/68060274)

```python
#零均值化
def zeroMean(dataMat):        
    meanVal=np.mean(dataMat,axis=0)     #按列求均值，即求各个特征的均值  
    newData=dataMat-meanVal  
    return newData,meanVal  

#选择主成份个数
 def percentage2n(eigVals,percentage):  
    sortArray=np.sort(eigVals)   #升序  
    sortArray=sortArray[-1::-1]  #逆转，即降序  
    arraySum=sum(sortArray)  
    tmpSum=0  
    num=0  
    for i in sortArray:  
        tmpSum+=i  
        num+=1  
        if tmpSum>=arraySum*percentage:  
            return num  

#PCA算法实现
def pca(dataMat,percentage=0.99):  
    newData,meanVal=zeroMean(dataMat)  
    covMat=np.cov(newData,rowvar=0)    #求协方差矩阵,return ndarray；若rowvar非0，一列代表一个样本，为0，一行代表一个样本  
    eigVals,eigVects=np.linalg.eig(np.mat(covMat))#求特征值和特征向量,特征向量是按列放的，即一列代表一个特征向量  
    n=percentage2n(eigVals,percentage)                 #要达到percent的方差百分比，需要前n个特征向量  
    eigValIndice=np.argsort(eigVals)            #对特征值从小到大排序  
    n_eigValIndice=eigValIndice[-1:-(n+1):-1]   #最大的n个特征值的下标  
    n_eigVect=eigVects[:,n_eigValIndice]        #最大的n个特征值对应的特征向量  
    lowDDataMat=newData*n_eigVect               #低维特征空间的数据  
    reconMat=(lowDDataMat*n_eigVect.T)+meanVal  #重构数据  
    return lowDDataMat,reconMat  
```
